## Comparing the result of different models

```python
# importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
```

## One Output Comparison

```python
df = pd.read_csv('results_max_proton_energy-1.csv')
df1 = pd.read_csv('metricsNNOneOutput50000Points.csv')
df2 = pd.read_csv('results.rf.csv')
df3 = pd.read_csv('results_max_proton_energy_Ascend_gpu.csv')
df4 =pd.read_csv('results_max_proton_energy_Pitzer_gpu.csv')


plt.plot(df['numpoints'],df['MSE'], marker="s", label='Decision-Tree')
plt.plot(df1['points'],df1['MSE'], marker="s",color= "orangered", label='Neural Network')
#plt.plot(df2['Number of Points'],df2['MSE'], marker="s", label='Random Forest')
#plt.plot(df3['numpoints'],df3['MSE'], marker="s", label='Ascend GPU Random Forest')
plt.plot(df4['numpoints'],df4['MSE'], marker="s",color= "Green", label='Random Forest')

plt.title(" Number of point vs MSE - 1 Output")
plt.ylim([0,0.010])
plt.legend(loc='upper right')
plt.xlabel("Number of points")
plt.ylabel("MSE ( $MeV^2$)")
plt.savefig('Decision-Tree-MSE-NN-rf -summer',dpi =600)
```

```python
plt.plot(df['numpoints'],df['ARE'], marker="s", label='Decision-Tree')
plt.plot(df1['points'],df1['Percent Error'], marker="s",color= "orangered", label='Neural Network')
#plt.plot(df2['Number of Points'],df2['ARE'], marker="s", label='Random Forest')
#plt.plot(df3['numpoints'],df3['ARE'], marker="s", label='Ascend GPU Random Forest')
plt.plot(df4['numpoints'],df4['ARE'], marker="s",color= "Green", label='Random Forest')

plt.title(" Number of point vs Avg Percent Error - 1 Output")
plt.ylim([8,17])
plt.legend(loc='upper right')
plt.xlabel("Number of points")
plt.ylabel("Percent Error (%)")
plt.savefig('Decision-Tree-AVG-NN-rf - summer',dpi =600)
```

```python
plt.plot(df['numpoints'],df['Elapsed Time (seconds)'], marker="s", label='Decision-Tree')
plt.plot(df1['points'],df1['yTime'], marker="s",color= "orangered", label='Neural Network')
#plt.plot(df2['Number of Points'],df2['Elapsed Time (seconds)'], marker="s", label='Random Forest')
#plt.plot(df3['numpoints'],df3['Elapsed Time (seconds)'], marker="s", label='Ascend GPU Random Forest')
plt.plot(df4['numpoints'],df4['Elapsed Time (seconds)'], marker="s",color= "Green", label='Random Forest')

plt.title(" Number of point vs Time - 1 Output")
plt.legend(loc='upper right')
plt.xlabel("Number of points")
plt.ylabel("Seconds")
plt.savefig('Decision-Tree-Time-NN-rf - summer',dpi =600)
```
