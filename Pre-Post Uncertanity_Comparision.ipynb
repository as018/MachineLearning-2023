```python
import pandas as pd
import matplotlib.pyplot as plt


df1 = pd.read_csv('datasets/Intensity/fuchs_v3_points_20000_noise_0.csv') 
df2 = pd.read_csv('datasets/Intensity/fuchs_v3_points_20000_noise_5.csv') 


df1 = df1.sort_values(by=['Max_Proton_Energy_(MeV)'])
df2 = df2.sort_values(by=['Max_Proton_Energy_(MeV)'])

# Calculate the relative error in max proton energy
df1['Relative_Error'] = (df2['Max_Proton_Energy_(MeV)'] - df1['Max_Proton_Energy_(MeV)']) / df1['Max_Proton_Energy_(MeV)']

# Plot a histogram of the relative errors
plt.figure(figsize=(10,6))
plt.hist(df1['Relative_Error'], bins=50, edgecolor='black',label='Intensity - Pre')

df1 = pd.read_csv('datasets/Energy/fuchs_v3_points_20000_noise_0.csv') 
df2 = pd.read_csv('datasets/Energy/fuchs_v3_points_20000_noise_5.csv') 

df1 = df1.sort_values(by=['Max_Proton_Energy_(MeV)'])
df2 = df2.sort_values(by=['Max_Proton_Energy_(MeV)'])

# Calculate the relative error in max proton energy
df1['Relative_Error'] = (df2['Max_Proton_Energy_(MeV)'] - df1['Max_Proton_Energy_(MeV)']) / df1['Max_Proton_Energy_(MeV)']


plt.hist(df1['Relative_Error'], bins=50, edgecolor='black',label="Energy - Post")

plt.xlabel('Relative Error')
plt.ylabel('Frequency')
plt.title('Histogram of Relative Error in Max Proton Energy')
plt.legend()
plt.grid(True)
plt.savefig("Histogram of Relative Error in Max Proton Energy")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('datasets/Intensity/fuchs_v3_points_20000_noise_0.csv') 
df2 = pd.read_csv('datasets/Intensity/fuchs_v3_points_20000_noise_5.csv') 

df1 = df1.sort_values(by=['Total_Proton_Energy_(MeV)'])
df2 = df2.sort_values(by=['Total_Proton_Energy_(MeV)'])

# Calculate the relative error in max proton energy
df1['Relative_Error'] = (df2['Total_Proton_Energy_(MeV)'] - df1['Total_Proton_Energy_(MeV)']) / df1['Total_Proton_Energy_(MeV)']

plt.figure(figsize=(4,2.5))
plt.hist(df1['Relative_Error'], bins=50, edgecolor='black',label='Input Error')

df1 = pd.read_csv('datasets/Energy/fuchs_v3_points_20000_noise_0.csv') 
df2 = pd.read_csv('datasets/Energy/fuchs_v3_points_20000_noise_5.csv') 

df1 = df1.sort_values(by=['Total_Proton_Energy_(MeV)'])
df2 = df2.sort_values(by=['Total_Proton_Energy_(MeV)'])

# Calculate the relative error in max proton energy
df1['Relative_Error'] = (df2['Total_Proton_Energy_(MeV)'] - df1['Total_Proton_Energy_(MeV)']) / df1['Total_Proton_Energy_(MeV)']


plt.hist(df1['Relative_Error'], bins=50, edgecolor='black',label="Output Error")

# Plot a histogram of the relative errors

plt.xlabel('Relative Error')
plt.ylabel('Frequency')
plt.legend()
plt.title('Relative Error in Total Proton Energy')
plt.tight_layout()
plt.grid(True)
plt.savefig('Histogram of Relative Error in Total_Proton_Energy_(MeV)',dpi =600)
plt.show()
```
